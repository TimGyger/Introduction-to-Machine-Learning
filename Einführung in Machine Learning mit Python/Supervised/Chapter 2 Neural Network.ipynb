# Artificial neural network (Deep Learning)

```python
import numpy as np
from scipy import sparse
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython import display
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X, y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y,
                                                    random_state = 0)

mlp = MLPClassifier(solver = 'lbfgs', random_state = 0).fit(X_train,y_train)

mglearn.plots.plot_2d_separator(mlp,X_train,fill = True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

We can define the number of hidden layers and their number of knots:

```python
# 3 layers with 10, 8 and 9 knots
mlp = MLPClassifier(solver = 'lbfgs', hidden_layer_sizes =[10,8,9],random_state = 0).fit(X_train,y_train)

mglearn.plots.plot_2d_separator(mlp,X_train,fill = True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

We can choose the activation function:

```python
mlp = MLPClassifier(solver = 'lbfgs', hidden_layer_sizes =[10,8,9],
                    activation = 'tanh',random_state = 0).fit(X_train,y_train)

mglearn.plots.plot_2d_separator(mlp,X_train,fill = True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

Again we can penalise with parameter alpha:

```python
mlp = MLPClassifier(solver = 'lbfgs', hidden_layer_sizes =[10,8,9],
                    activation = 'tanh',random_state = 0, alpha = 1).fit(X_train,y_train)

mglearn.plots.plot_2d_separator(mlp,X_train,fill = True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

To improve our results we have to scale our features:

```python
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis = 0)

X_train_scaled = (X_train - mean_on_train)/std_on_train
X_test_scaled = (X_test - mean_on_train)/std_on_train

mlp = MLPClassifier(random_state = 0, alpha = 0.1, max_iter = 1000).fit(X_train_scaled,y_train)

mglearn.plots.plot_2d_separator(mlp,X_train_scaled,fill = True, alpha=.3)
mglearn.discrete_scatter(X_train_scaled[:,0],X_train_scaled[:,1],y_train)
```
