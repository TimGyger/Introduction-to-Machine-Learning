# Manifold Learning

By transforming the data we try to visualise, comprimise and represent the data in a way that makes it possible to extract information. Manifold Learning Algos are aiming for the visualisation and therefore generate rarely more than two features.

```python
import numpy as np
from scipy import sparse
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython import display
```

## t-SNE

t-SNE tries to represent the data in 2D in a way that the distances between points are nearly preserved. It starts with a random distribution in 2D and tries to position points nearer if they are near in the space of all features and so on. Therefore we can extract information about the relation of neighbors.

```python
from sklearn.datasets import load_digits
digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10,5),
                        subplot_kw={'xticks': (), 'yticks': ()})

for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

### PCA:

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(digits.data)

digits_pca = pca.transform(digits.data)

plt.figure(figsize=(15,15))
plt.xlim(digits_pca[:,0].min(), digits_pca[:,0].max())
plt.ylim(digits_pca[:,1].min(), digits_pca[:,1].max())

for i in range(len(digits.data)):
    plt.text(digits_pca[i,0],digits_pca[i,1], str(digits.target[i]), fontdict ={'weight': 'bold', 'size': 12})
    
plt.xlabel("1. Component")  
plt.ylabel("2. Component")  
```

```python
# t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(random_state=0)
digits_tsne = tsne.fit_transform(digits.data)

plt.figure(figsize=(15,15))
plt.xlim(digits_tsne[:,0].min(), digits_tsne[:,0].max())
plt.ylim(digits_tsne[:,1].min(), digits_tsne[:,1].max())

for i in range(len(digits.data)):
    plt.text(digits_tsne[i,0],digits_tsne[i,1], str(digits.target[i]), fontdict ={'weight': 'bold', 'size': 12})
    
plt.xlabel("1. t-SNE Feature")  
plt.ylabel("2. t-SNE Feature")  
```
